```python
import os
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from ultralytics import YOLO
from sklearn.metrics import classification_report, confusion_matrix
```

```python
MODEL_PATH = "/content/drive/MyDrive/Colab Notebooks/epoch45.pt"
model = YOLO(MODEL_PATH)
```

```python
TEST_DIR = "/content/drive/MyDrive/Colab Notebooks/Test images_each 20"
```

```python
TEST_DIR = "/content/drive/MyDrive/Colab Notebooks/Test images_each 20"

class_names = sorted(os.listdir(TEST_DIR))
class_to_idx = {cls: i for i, cls in enumerate(class_names)}

y_true = []
y_pred = []

for cls in class_names:
    cls_path = os.path.join(TEST_DIR, cls)

    for img_name in os.listdir(cls_path):
        img_path = os.path.join(cls_path, img_name)

        result = model(img_path, verbose=False)[0]

        # If NO detection → skip
        if result.boxes is None or len(result.boxes) == 0:
            continue

        # Take highest confidence detection
        confs = result.boxes.conf.cpu().numpy()
        classes = result.boxes.cls.cpu().numpy().astype(int)

        best_idx = np.argmax(confs)
        pred_class = classes[best_idx]

        y_true.append(class_to_idx[cls])
        y_pred.append(pred_class)
```

```python
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=class_names,
    yticklabels=class_names
)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
print(
    classification_report(
        y_true,
        y_pred,
        target_names=class_names
    )
)
```
